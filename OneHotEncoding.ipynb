## 1. OneHotEncoding using Pandas

```python
df['owner'].value_counts()
```

```python
pd.get_dummies(df, columns=['fuel','owner'])
# result : we get 12 col, fuel is converted to 4 different cols, owner is also converted to 5 different cols
```

## 2. K-1 OneHotEncoding (removing one col)

```python
pd.get_dummies(df, columns=['fuel','owner'], drop_first=True)
# Now we have 10 col since 1 col is removed from ach encoding cols (fuel and owner)
```

## 4. OneHotEncoding on Top Categories

## used when when we have some categories with very high frequency while others with very low frequency

```python
df['brand'].value_counts()
```

```python
df['brand'].nunique()
threshold = 100
count = df['brand'].value_counts()
```

```python
repl = count[count <= threshold].index
```

```python
pd.get_dummies(df['brand'].replace(repl,'uncommon'))
```
